# 图表的常用设置

## 基本绘图 plot 函数

语法: `plt.plot(x, y, format_string, **kwargs)`

参数：

  - x: x轴数据
  - y: y轴数据
  - format_string: 控制曲线格式的字符串，包括颜色、线条样式和标记样式。

```python
import matplotlib.pyplot as plt
# 折线图
x = range(1, 15, 1)
y = range(1, 42, 3)
plt.plot(x, y)
plt.show()
```

# 常用图标的绘制

## 绘制多个子图表

Matplotlib提供了三种方法在一张图上绘制多个子图表：`subplot函数`, `subplots函数`, `add_subplot函数`。

#### subplot 函数

subplot 函数直接指定划分方式和位置，将一个绘图区划分为n个子图，每个subplot函数只能绘制一个子图。

例如，绘制一个 2x3 的区域, subplot(2,3,3), 将画布分成 2行3列 在 第3个区域中绘制；如果行列的值都小于10，可以缩写为一个整数，例如，subplot(223)。

```python
import matplotlib.pyplot as plt
for i in range(1, 7):
    plt.subplot(2, 3, i)
plt.show()
```
